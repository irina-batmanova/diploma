```python
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
import pandas as pd
```

```python
domain = "https://www.kinopoisk.ru"
```

```python
base_url = domain + "/top/lists/"
```

```python
response = requests.get(base_url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
topic_list_block = soup.find("div", {"class": "list_main js-rum-hero"})
```

```python
topic_url_list = topic_list_block.findAll("li", {"class": " userFill0"})
```

```python
hrefs = []
for url in topic_url_list:
    hrefs.append(url.find("a")["href"])
```

```python
print(hrefs[1])
```

```python
response_dict = {"response_text": [], "sentiment": []}
```

```python
for href in tqdm(hrefs[1:]):
    request_url = domain + href
    film_hrefs = []
    response = requests.get(request_url)
    filmsoup = BeautifulSoup(response.text, 'html.parser')
    film_block = soup.find("table", {"class": "ten_items js-rum-hero"})
    film_list = film_block.findAll("tr", {"class": " js-film-list-item"})
    for film in film_list:
        film_request_url = domain + "/film/" + film["id"].split("_")[1] + "/reviews/?ord=rating"
        film_response = requests.get(film_request_url)
        soup_per_film = BeautifulSoup(film_response.text, 'html.parser')
        good_responses = soup_per_film.findAll("div", {"class": "response good"})
        for resp in good_responses:
            resp_body = resp.find("span", {"class": "_reachbanner_"})
            response_dict["response_text"].append(resp_body.text)
            response_dict["sentiment"].append(1)
        bad_responses = soup_per_film.findAll("div", {"class": "response bad"})
        for resp in bad_responses:
            resp_body = resp.find("span", {"class": "_reachbanner_"})
            response_dict["response_text"].append(resp_body.text)
            response_dict["sentiment"].append(-1)
```

```python
print(len(response_dict))
```

```python
df = pd.DataFrame(response_dict)
```
